```python
corpus = [ 'I have finished the NLP exam today',' My friends and I studied the NLP together yesterday','Because I studied hard, I passed the exam, and I was so happy'
 ]
```

```python
import string
words_set = set()

for doc in  corpus:
    words = ''.join(char.lower() if char not in string.punctuation else ' ' for char in doc).split()
    words_set.update(words)

print('Number of words in the corpus:',len(words_set))
print('The words in the corpus: \n', words_set)
```

```python
import pandas as pd
import numpy as np
import string
from fractions import Fraction

n_docs = len(corpus)         # Number of documents in the corpus
n_words_set = len(words_set) # Number of unique words in the set

df_tf = pd.DataFrame(np.zeros((n_docs, n_words_set), dtype=object), columns=list(words_set))

# Compute Term Frequency (TF)
for i in range(n_docs):
    words = ''.join(char.lower() if char not in string.punctuation else ' ' for char in corpus[i]).split() # Remove punctuation
    for w in words:
        w_lower = w.lower()  # Convert word to lowercase
        # Use Fraction to represent TF as a fraction
        df_tf.at[i, w_lower] += Fraction(1, len(words))

# Display the DataFrame with TF as fractions
print("TF DataFrame:")
print(df_tf)
```

```python
import pandas as pd
import numpy as np
import string

n_docs = len(corpus)         # Number of documents in the corpus
n_words_set = len(words_set) # Number of unique words in the set

df_tf = pd.DataFrame(np.zeros((n_docs, n_words_set)), columns=list(words_set))

# Compute Term Frequency (TF)
for i in range(n_docs):
    words = ''.join(char.lower() if char not in string.punctuation else ' ' for char in corpus[i]).split() # Remove punctuation
    for w in words:
        w_lower = w.lower()  # Convert word to lowercase
        df_tf.at[i, w_lower] += (1 / len(words))

df_tf
```

```python
df_idf = pd.DataFrame(index=['IDF'], columns=df_tf.columns)
for col in df_idf.columns:
    # Count of documents in the corpus that contain this word
    doc_count = sum(1 for value in df_tf[col] if value > 0)

    # Calculate IDF and store in the DataFrame
    df_idf[col] = np.log10(n_docs / (doc_count))

df_idf
```

```python
# Compute TF-IDF
df_tfidf = df_tf * df_idf.loc['IDF']

# Display the TF-IDF DataFrame
print("TF-IDF DataFrame:")
print(df_tfidf)
```

```python
average_tfidf = df_tfidf.mean(axis=1)

# Display the average TF-IDF values
print("Average TF-IDF values:")
print(average_tfidf)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Create a CountVectorizer
corpus_text = [' '.join(''.join(char.lower() if char not in string.punctuation else ' ' for char in doc).split()) for doc in corpus]
vectorizer = CountVectorizer(lowercase=True)

# Fit and transform the corpus text
X = vectorizer.fit_transform(corpus_text)

# Create a DataFrame from the CountVectorizer output
df_bow = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
df_bow
```

```python
# Calculate the sum of each column (word frequency across all documents)
sum_per_word = df_bow.sum(axis=0)

# Display the sum of each column
print("Sum of Each Column (Word Frequency Across All Documents):")
print(sum_per_word)
```
